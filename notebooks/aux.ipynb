```python
# !pip install matplotlib
import random
import matplotlib.pyplot as plt

def monte_carlo_visualization(n_points: int):
    inside_x = []
    inside_y = []
    outside_x = []
    outside_y = []
    total_inside = 0

    for _ in range(n_points):
        x, y = random.uniform(-1, 1), random.uniform(-1, 1)
        if x**2 + y**2 <= 1:
            inside_x.append(x)
            inside_y.append(y)
            total_inside += 1
        else:
            outside_x.append(x)
            outside_y.append(y)

    pi_estimate = (total_inside / n_points) * 4

    fig, ax = plt.subplots(figsize=(6, 6))

    circle = plt.Circle((0, 0), 1, color='blue', fill=False, linewidth=2)
    ax.add_patch(circle)

    ax.scatter(inside_x, inside_y, color='green', s=5, label="Dentro do crculo")
    ax.scatter(outside_x, outside_y, color='red', s=5, label="Fora do crculo")

    ax.set_xlim(-1, 1)
    ax.set_ylim(-1, 1)
    ax.set_aspect('equal')
    ax.set_title(f"Mtodo de Monte Carlo - {n_points} pontos\nPontos dentro do Crculo = {total_inside}\nEstimativa de   {pi_estimate:.5f}")
    plt.show()

    print(f"Total de pontos: {n_points}")
    print(f"Dentro do crculo: {total_inside}")
    print(f"Estimativa de : {pi_estimate:.5f}")

monte_carlo_visualization(100000)
```
